```python
import xarray as xr
import glob
import os
```

```python
path = "./data"
nc_files = glob.glob(os.path.join(path, "*.nc"))

datasets = [print(f) for f in nc_files]
datasets = [xr.open_dataset(f, engine='netcdf4') for f in nc_files]

# Combine the datasets along the time dimension
combined_ds = xr.concat(datasets, dim='time')

combined_ds
```
